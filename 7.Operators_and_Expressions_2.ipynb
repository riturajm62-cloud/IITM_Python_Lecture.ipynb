```python
# Arithematic operators(+, -, *, /, //, %, **)
print('Addition', 2 + 3)
print('Subtraction', 9 - 2)
print("Multiplication", 15 * 13)
print("Division", 136 / 23)
print("floor division", 7 // 2)
print("Modulus", 7 % 3)
print("exponentiation", 2 ** 3)

```

```python
# Relational operators(>, <, >=, <=, ==, !=)
print(5 > 10)
print(5 < 100)
print(5 >= 5)
print(5 <= 5)
print(5 == 50)
print(5 == 5)
print(5 != 50)
print(5 != 5)
      
```

```python
# Logical operators(and, or, not)
print(True and True)
print(True and False)
print(False and True)
print(False and False)
```

```python
# Logical operators(and, or, not)
print(True or True)
print(True or False)
print(False or True)
print(False or False)
```

```python
# Logical operators(and, or, not)
print(not (True))
print(not False)

```
