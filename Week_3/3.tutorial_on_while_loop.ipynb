```python
#Find factorial of given number
num =int(input('Enter a number:'))
fact = 1
if (num < 0):
    print('Not defined')
else:
    while(num >0):
        fact = fact*num
        num = num-1
    print(fact)
```

```python
# Reverse the digits in the given number
num = int(input('Enter a number:'))
absnum = abs(num)
rev = absnum % 10
absnum = absnum // 10
while(absnum > 0):
    r = absnum % 10
    absnum = absnum // 10
    rev = rev * 10 + r
if (num > 0):
    print(rev)
else:
    print(rev - 2 * rev)
```

```python
# Find entered number is palindrome or not
num = int(input('Enter a number:'))
absnum = abs(num)
rev = absnum % 10
absnum = absnum // 10
while(absnum > 0):
    r = absnum % 10
    absnum = absnum // 10
    rev = rev * 10 + r
if (num < 0):
    rev =  rev - 2 * rev
if (num == rev):
    print('Palindrome')
else:
    print('Not a Palindrome')
```
