```python
#number of iteration is not known to us
num =int(input('Enter a number:'))
fact = 1
if (num < 0):
    print('Not defined')
else:
    while(num >0):
        fact = fact*num
        num = num-1
    print(fact)
```

```python
#number of iteration is  known to us
num =int(input('Enter a number:'))
fact = 1
if (num < 0):
    print('Not defined')
else:
    for i in range(num, 1, -1):
        fact = fact * i
    print(fact)
```
