```python
#positional arguments
def add(a,b,c):
    return (a+b-c)

print(add(20,30,40))
```

```python
def add(c,b,a):
    return (a+b-c)

print(add(20,30,40))
```

```python
#keyword argument
def add(c,b,a):
    return(a+b-c)

print(add(a=20,b=30,c=40))
```

```python
#default argument
def add(c, a=20,b=30):
    return (a+b-c)

print(add(40))
print(add(10,40))
print(add(10,50,60))
```

```python
#mixed argument
def add(c, a=20,b=30):
    return (a+b-c)

print(add(40))
print(add(10,40))
print(add(10,a=400,b=60))
```

```python
#mixed argument
def add(c, a=20,b=30):
    print(a+b-c)

print(add(40))
print(add(10,40))
print(add(10,a=400,b=60))
```
