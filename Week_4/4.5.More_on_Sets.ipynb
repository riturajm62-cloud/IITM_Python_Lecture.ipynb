```python
st={1,2,3,4,5}
#set dos not allow duplicate value
print(st)
#set is an unordered entity.
#set object is not subscriptable.
#set is simply a collection of value
```

```python
'''set is peculiar,set itself is mutable, but every value inside set has to be immutable and hashable,
which means you can add value to the set but whatever values which we add , those value has to be immutable.'''
```

```python
#set operations
A={1,3,5}
B={1,2,3,4,5}
print(A.issubset(B))
```

```python
print(A.issuperset(B))
```

```python
A={1,2,3}
B={3,4,5}
c1=A.union(B)
c2=A | B
print(c1,c2)
```

```python
A={1,2,3}
B={3,4,5}
c1=A.intersection(B)
c2=A & B
print(c1,c2)
```

```python
A={1,2,3}
B={3,4,5}
c1=A.difference(B)
c2=A - B
print(c1,c2)
```
