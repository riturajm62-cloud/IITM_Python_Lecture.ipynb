```python
# List Comprehension
fruits=["mango", "apple","orange","pear","guava","banana"]
newlist = []
for fruit in fruits:
    if 'n' in fruit:
        newlist.append(fruit.capitalize())
print(newlist)
```

```python
# List Comprehension
fruits=["mango", "apple","orange","pear","guava","banana"]

newlist = [fruit.capitalize() for fruit in fruits if 'n' in fruit ]
print(newlist)
```
