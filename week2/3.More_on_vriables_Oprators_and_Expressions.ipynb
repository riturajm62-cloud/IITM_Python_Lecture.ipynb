```python
#We must start a variable name with alhabet or with underscore and not with number.
# Variables names are case sensitive.
```

```python
#Multiple assignment
x,y=1,2
print(x,y)

```

```python
x=y=z=10
print(x,y,z)
```

```python
x,y=1,2
print(x,y)
x,y=y,x
print(x,y)
```

```python
# Deleting a variable
x=10
print(x)
del x
print(x)
```

```python
# Shorthand operators help us to remove a repetitive use of same variable name in the same statement.
count=0
count=count+1
count=count+1
print(count)

# Shorthand operators help us to remove a repetitive use of same variable name in the same statement.
count=0
count +=1
count +=1
print(count)
```

```python
count=2
count *=2
count /=4
print(count)
```

```python
# Special operator called in operator
print('alpha' in 'A variable name can only contain alpha-numeric characters and underscore')
print('alpha' in 'A variable name must start ith a letter or the underscore character')
```

```python
# Chaining operators.
x=5
print(1<x<10)
print(10<x<20)
print(x<10<x*10<100)
print(10>x<=9)
print(5==x>4)
```
