```python
x="You are my sunshine"
print(x.lower())
print(x.upper())
print(x.capitalize())
print(x.title())
print(x.swapcase())
print(x.islower())
print(x.isupper())
print(x.istitle())
```

```python
x='123'
print(x.isdigit())
print(x.isalpha())
print(x.isalnum())
```

```python
x="Who are you"
print(x.strip(','))
print(x.lstrip('-'))
print(x.rstrip('-'))
```

```python
x="Ritu Raj"
print(x.startswith('R'))
print(x.endswith('j'))
print(x.count('R'))
print(x.index('j'))
print(x.replace('j','z'))
```
