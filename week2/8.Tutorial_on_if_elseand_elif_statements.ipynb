```python
# To find given number is even or odd.
num = int(input('Enter a number: '))
if (num % 2==0):
    print('Even')
else:
    print('Odd')
```

```python
# chek last digit of given number ends with 0 or 5
num = int(input('Enter a number: '))
if (num%5==0):
    if (num%10==0):
        print('0')
    else:
        print('5')
else:
    print('Other')
```

```python
# Finding grade based on given marks and given condition
marks = int(input('Enter a marks: '))
if (marks >=0 and marks <=100):
    if (marks>=90):
        print('A')
    if (marks >=80 and marks <90):
        print('B')
    if (marks >=70 and marks <80):
        print('C')
    if (marks >=60 and marks <70):
        print('D')
    if (marks < 60):
        print('E')
else:
    print('Invalid input')

```
